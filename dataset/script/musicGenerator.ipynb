```python
import wave
def get_duration_wave(file_path):
   with wave.open(file_path, 'r') as audio_file:
      frame_rate = audio_file.getframerate()
      n_frames = audio_file.getnframes()
      duration = n_frames / float(frame_rate)
      return duration
file_path = '../audio/test1.wav'
duration = get_duration_wave(file_path)
print(f"Duration: {duration:.2f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define sample rate and time array
sample_rate = 44100
t = np.linspace(0, 1, sample_rate, endpoint=False)

# Frequencies of two notes
f1 = 440  # A4
f2 = 523.25  # C5

# Generate the waveforms for each note
note1 = np.sin(2 * np.pi * f1 * t)
note2 = np.sin(2 * np.pi * f2 * t)

# Combine the notes
combined = note1 + note2

# Plot the waveforms
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(t[:1000], note1[:1000])
plt.title('Note 1 Waveform')
plt.subplot(1, 3, 2)
plt.plot(t[:1000], note2[:1000])
plt.title('Note 2 Waveform')
plt.subplot(1, 3, 3)
plt.plot(t[:1000], combined[:1000])
plt.title('Combined Waveform')
plt.tight_layout()
plt.show()
```
